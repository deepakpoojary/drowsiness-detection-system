```python
import cv2
import dlib
import numpy as np
from scipy.spatial import distance as dist
from pygame import mixer
import time
mixer.init()
sound = mixer.Sound('alarm.wav')

# Load face detector and landmark predictor
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")

def mouth_aspect_ratio(mouth):
    A=dist.euclidean(mouth[0],mouth[4])
    B=dist.euclidean(mouth[2],mouth[6])
    return (B/A)

# Calculate eye aspect ratio (EAR) to detect drowsiness
def eye_aspect_ratio(eye):
    A = dist.euclidean(eye[1], eye[5])
    B = dist.euclidean(eye[2], eye[4])
    C = dist.euclidean(eye[0], eye[3])
    ear = (A + B) / (2.0 * C)
    return ear
```

```python
import os
from sklearn.metrics import classification_report,accuracy_score
import matplotlib.pyplot as plt
lis=[]
test=[]
entries = os.listdir('ClosedFace/')
for i in entries:
    if i=='Thumbs.db':
        continue
    lis.append(1)
    test.append(0)
    img=cv2.imread('ClosedFace//'+i)
    
    img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    
    faces = detector(img, 0)
 
    
    for face in faces:
            # Detect landmarks on the face
            landmarks = predictor(img, face)
            
        

            # Extract left and right eye coordinates
           
           
            left_eye = []
            
            for i in range(36, 42):
                left_eye.append((landmarks.part(i).x, landmarks.part(i).y))
            right_eye = []
            for i in range(42, 48):
                right_eye.append((landmarks.part(i).x, landmarks.part(i).y))
            
        
        
            # Calculate EAR for left and right eyes
            left_ear = eye_aspect_ratio(left_eye)
            right_ear = eye_aspect_ratio(right_eye)
       
            # Calculate average EAR
            ear = (left_ear + right_ear) / 2.0
            if(ear<0.25):
                test.pop()
                test.append(1)    

print(classification_report(lis,test))
acc=accuracy_score(lis,test)
    
```

```python
# Load video stream
cap = cv2.VideoCapture(0)

# Initialize variables
frame_counter = 0
frame_counter1=0
ALARM_ON = False

while True:
    ret, frame = cap.read()

    # Convert image to grayscale because It helps in simplifying algorithms and as well 
    # eliminates the complexities related to computational requirements. It makes room for
    # easier learning for those who are new to image processing. This is because grayscale 
    # compressors an image to its barest minimum pixel.
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect faces in the image
    faces = detector(gray, 0)
    for face in faces:
        # Detect landmarks on the face
        landmarks = predictor(gray, face)
       # Extract left and right eye coordinates
        left_eye = []
        mouth=[]
        
        for i in range(36, 42):
            left_eye.append((landmarks.part(i).x, landmarks.part(i).y))
        right_eye = []
        for i in range(42, 48):
            right_eye.append((landmarks.part(i).x, landmarks.part(i).y))
        for i in range(60,68):
            mouth.append((landmarks.part(i).x,landmarks.part(i).y))
        # Calculate EAR for left and right eyes
        left_ear = eye_aspect_ratio(left_eye)
        right_ear = eye_aspect_ratio(right_eye)
        mouth_ar=mouth_aspect_ratio(mouth)
        if(mouth_ar>0.25):
            frame_counter1 += 1
            if frame_counter1 >= 20:
                cv2.putText(frame, "Yawning", (150, 20), cv2.CALIB_CB_NORMALIZE_IMAGE, 0.75, (149, 225, 255), 1)
       # Calculate average EAR
        ear = (left_ear + right_ear) / 2.0
        # Detect drowsiness if EAR is below threshold for a certain number of frames
        if ear < 0.25:
            
            frame_counter += 1
            if frame_counter >= 30:
                cv2.putText(frame, "Drowsiness Detected", (20, 250), cv2.CALIB_CB_NORMALIZE_IMAGE, 0.75, (149, 225, 255), 1)
                
                
                if not ALARM_ON:
                    ALARM_ON = True
                    try:
                     sound.play()
                      # time.sleep(0.05)
                     
                    except: # isplaying = False
                     pass
                     # Start alarm or any other alert mechanism
        else:
            frame_counter = 0
            ALARM_ON = False
            sound.stop()
        # # Draw eyes and EAR on the frame
        # cv2.drawContours(frame, [np.array(left_eye)], -1, (0, 255, 0), 1)
        # cv2.drawContours(frame, [np.array(right_eye)], -1, (0, 255, 0), 1)
        cv2.putText(frame, "EAR: {:.2f}".format(ear), (20, 20), cv2.CALIB_CB_NORMALIZE_IMAGE, 0.75, (149, 49, 255), 1)
        # Show the frame
    cv2.imshow("Drowsiness Detection", frame)
      
    if cv2.waitKey(1) == ord('a'):
        break
    # Release the video stream and close all windows
cap.release()
cv2.destroyAllWindows()
```
